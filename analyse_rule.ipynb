```python
import pandas as pd
from metric import h
import os
def get_related_commit(add_lib, rem_lib):
    pd.set_option('display.width', None)
    path = f'cache/{rem_lib}.xlsx'
    df = pd.read_excel(path)
    df = df[(df['add_lib']==add_lib) & (df['rem_lib']==rem_lib)]
    df = df.drop(columns=['add_commit', 'rem_commit'])
    # related_df = pd.DataFrame(columns=['rem_repo', 'add_lib', 'rem_lib', 'add_message', 'rem_message'])
    related_df = pd.DataFrame(columns=['rem_repo', 'add_message', 'rem_message'])
    for s in df.values:
        rem_repo = s[0]
        add_message = s[3]
        rem_message = s[4]
        add_lib = s[1]
        rem_lib = s[2]
        if h(add_message, rem_message, add_lib, rem_lib):
            data = {
                'rem_repo': rem_repo,
                # 'add_lib': add_lib,
                # 'rem_lib': rem_lib,
                'add_message': add_message,
                'rem_message': rem_message
            }
            related_df = related_df.append(data, ignore_index=True)
    save_path = f'related_commit/{rem_lib}'
    if not os.path.exists(save_path):
        os.mkdir(f'related_commit/{rem_lib}')
    related_df.to_excel(save_path + f'/{rem_lib}_to_{add_lib}.xlsx', index=False)
    return related_df

add_lib = 'authlib'
rem_lib = 'httplib2'
related_commit = get_related_commit(add_lib, rem_lib)
print(len(related_commit))
print(related_commit)
```
